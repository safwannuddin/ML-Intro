### mathplotlib example Workflow 

```python
%matplotlib inline 
import matplotlib.pyplot as plt 

#1preparing the data
x=[1,2,3,4,5]
y=[10,30,40,50,60]

#2 set up the plot 
fig,ax=plt.subplots(figsize=(10,10))

#3 plot data
ax.plot(x,y)

#4 Custamise the data 

ax.set(title="Simple plot",
       xlabel="X-axis",
       ylabel="Y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Use a dictionary instead of a set
nuts_butter = {
    "Almonds_butter": 20,
    "Cashew_butter": 15,
    "Peanut_butter": 30
}

fig, ax = plt.subplots()
ax.bar(nuts_butter.keys(), nuts_butter.values())  # 

plt.show()
```

```python
import matplotlib.pyplot as plt

nuts_butter = {
    "Almonds_butter": 20,
    "Cashew_butter": 15,
    "Peanut_butter": 30
}

fig, ax = plt.subplots()
ax.bar(list(nuts_butter.keys()), list(nuts_butter.values()))  #  No semicolon needed
plt.show()
```
